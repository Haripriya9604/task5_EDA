```python
%matplotlib inline
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

sns.set(style="whitegrid")
plt.rcParams['figure.dpi'] = 100
```

```python
file_path = 'train.csv'
if not os.path.exists(file_path):
    raise FileNotFoundError(f"train.csv not found in working dir: {os.getcwd()}")
df = pd.read_csv(file_path)
print("Loaded:", file_path)
print("Shape:", df.shape)
display(df.head())
```

```python
print("==== Info ====")
display(df.info())
print("\n==== Numeric summary ====")
display(df.describe(include='all').T)
print("\n==== Missing values (before) ====")
display(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
if 'Embarked' in df.columns:
    df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.drop(columns=['Cabin'], inplace=True, errors='ignore')

print("Missing values (after):")
display(df.isnull().sum())
```

```python
df['Survived'] = df['Survived'].astype(int)
df['Pclass'] = df['Pclass'].astype(int)
df['Sex'] = df['Sex'].astype('category')
if 'Embarked' in df.columns:
    df['Embarked'] = df['Embarked'].astype('category')

print("Survival distribution (counts & proportion):")
display(df['Survived'].value_counts().rename_axis('Survived').reset_index(name='Count'))
display(df['Survived'].value_counts(normalize=True).rename_axis('Survived').reset_index(name='Proportion'))

print("\nSurvival mean by Sex and Pclass:")
display(df.groupby(['Sex','Pclass'])['Survived'].mean().unstack())
```

```python
num_cols = ['Age','Fare']
for col in num_cols:
    plt.figure(figsize=(7,4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.show()
    plt.figure(figsize=(7,2.5))
    sns.boxplot(x=df[col].dropna())
    plt.title(f'Boxplot of {col}')
    plt.show()

cat_cols = ['Sex','Pclass'] + (['Embarked'] if 'Embarked' in df.columns else [])
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df, order=sorted(df[col].unique()) if col=='Pclass' else None)
    plt.title(f'Count of {col}')
    plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(1,3,1)
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')

plt.subplot(1,3,2)
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Pclass')

plt.subplot(1,3,3)
if 'Embarked' in df.columns:
    sns.countplot(x='Embarked', hue='Survived', data=df)
    plt.title('Survival by Embarked')
else:
    plt.text(0.5,0.5,"Embarked not present",ha='center',va='center')
    plt.axis('off')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age by Survived')

plt.subplot(1,2,2)
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare by Survived')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, alpha=0.6)
plt.title('Fare vs Age (colored by Survived)')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.show()

plt.figure(figsize=(7,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Numeric correlation matrix')
plt.show()

pair_cols = ['Age','Fare','SibSp','Parch','Survived']
pair_sample = df.dropna(subset=pair_cols).sample(n=min(500, len(df)), random_state=1)
sns.pairplot(pair_sample[pair_cols], hue='Survived', diag_kind='hist', corner=True)
plt.suptitle('Pairplot (sample)', y=1.02)
plt.show()
```

```python
print("Survival rates by Sex and Pclass:")
display(df.groupby(['Sex','Pclass'])['Survived'].mean().unstack())

if 'Embarked' in df.columns:
    print("\nSurvival counts by Embarked:")
    display(pd.crosstab(df['Embarked'], df['Survived'], margins=True))

print("\nStarter observations:")
print("1. Females show higher survival proportions than males.")
print("2. 1st class passengers tended to survive more often than 3rd class.")
print("3. Younger age and higher fare appear associated with higher survival rates.")
print("4. Use these findings to guide feature engineering (e.g., title extraction from names, fare bins, age groups).")
```
